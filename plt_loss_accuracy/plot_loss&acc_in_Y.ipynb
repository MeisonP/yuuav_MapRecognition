```python
#coding:utf-8
#20180818 
import matplotlib.pyplot as plt
from matplotlib.pyplot import plot,savefig
import numpy as np
import re


#############prepare the data 
acc_file="20180818_acc.txt" 
loss_file="20180818_loss.txt"
loss_x,loss_y,acc_x,acc_y=([],[],[],[])

def fdata(filename,appended_axis,start,end,type_):
    with open(filename,'r') as f:
        while True:
            line=f.readline()
            if not line:
                break
            result=re.split("{}|{}".format(start,end),line)[1]
            if type_=='int':
                result=int(result)
            if type_=='float':
                result=float(result)
            appended_axis.append(result)
            
fdata(acc_file,acc_x,'Iteration' ,', Testing','int')
fdata(acc_file,acc_y,'detection_eval = ',"\n",'float')    
fdata(loss_file,loss_x,'Iteration ',', loss = ','int')
fdata(loss_file,loss_y,'mbox_loss = ','\n','float')  



##############plot

# an empty figure with no axes
fig = plt.figure()

#subplot,this two are same: fig.add_subplot() or plt.subplot() 
ax1 = fig.add_subplot(111)

#title and axis label
ax1.set_title("Loss&Acc")
ax1.set_ylabel("Loss")
ax1.set_ylim([0,20])

##plot
ax1.plot(loss_x,loss_y,label='loss')

# add the plot2
ax2 = ax1.twinx() #important
ax2.set_ylabel("mAP")
ax2.set_ylim([0,0.5])

#ax2.plot(acc_x,acc_y,label='acc',color='r',linewidth=4,linestyle='-',markersize = 15, marker='s',)
area=[]
N=np.array(acc_y).shape[0]
for i in range(0,N):
    area.append((np.array(acc_y)[i]*35)**3)
plt.scatter(acc_x, acc_y, s=area, color='b', alpha=0.6)
#show the plot and save
plot()#show the plot will contain the axis , if using plt.show() then will contain no axis
savefig("train_result.png")
```
